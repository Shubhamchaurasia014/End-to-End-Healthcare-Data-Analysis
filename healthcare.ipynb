## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## Load the data

```python
# Load the data and read top 5 rows
df = pd.read_csv(r"D:\Shubham\Healthcare\hospital_final.csv")
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

## Data cleaning

```python
# convert date columns
df['appointment_date'] = pd.to_datetime(df['appointment_date'])
df['treatment_date'] = pd.to_datetime(df['treatment_date'])
```

## EDA

```python
df.groupby(['payment_method'])['amount'].mean().plot(kind='line',  figsize=(6,4))
```

```python
df.groupby(['treatment_type'])['amount'].sum().sort_values().plot(kind='barh')
```

```python
df['month'] = df['appointment_date'].dt.month
df.groupby(['month'])['amount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
df.groupby(['patient_id'])['amount'].sum().sort_values(ascending=False).head().plot(kind='bar')
```
